```python
def run_c3_mmfd(text=None, video_path=None):
    """
    Runs the full C³-MMFD framework on given inputs
    and prints a clean, human-readable final decision.
    """

    print("\n" + "=" * 60)
    print("Running C³-MMFD Framework")
    print("=" * 60)

    # ----------------------------
    # 1️⃣ TEXT MODULE
    # ----------------------------
    if text:
        print("\n[1] TEXT MODULE")
        text_out = text_verify(text)
        print("Text result:")
        print(f"  Label      : {text_out['label']}")
        print(f"  Confidence : {text_out['confidence']:.2f}")
        print(f"  Reason     : {text_out['reason']}")
    else:
        text_out = None
        print("\n[1] TEXT MODULE: Skipped")

    # ----------------------------
    # 2️⃣ AUDIO–VIDEO MODULE
    # ----------------------------
    if video_path:
        print("\n[2] AUDIO–VIDEO MODULE")
        av_out = multimodal_fusion(video_path)

        print("AV result:")
        print(f"  Video label : {av_out['video_label']}")
        print(f"  Video score : {av_out['video_spoof_score']:.2f}")
        print(f"  Final label : {av_out['label']}")

        if av_out.get("low_quality_video", False):
            print("  ⚠ Low-quality video detected (confidence downgraded)")

        if av_out.get("audio_used", False):
            print("  Audio used  : Yes")
        else:
            print("  Audio used  : No")
    else:
        av_out = None
        print("\n[2] AUDIO–VIDEO MODULE: Skipped")

    # ----------------------------
    # 3️⃣ FINAL DECISION LAYER
    # ----------------------------
    decision = final_decision(
        text_out=text_out,
        av_out=av_out
    )

    # ----------------------------
    # ✅ CLEAN FINAL OUTPUT
    # ----------------------------
    print("\n" + "-" * 60)
    print("FINAL VERDICT (C³-MMFD)")
    print("-" * 60)

    print(f"Label           : {decision['final_label']}")
    print(f"Decision type   : {decision['decision_type']}")
    print(f"Trusted source  : {decision['trusted_modality']}")
    print(f"Explanation     : {decision['explanation']}")

    print("\nReliability:")
    rel = decision["reliability"]
    print(f"  Text reliable : {rel['text_reliable']}")
    print(f"  AV reliable   : {rel['av_reliable']}")
    print(f"  Notes         : {rel['notes']}")

    print("\nConsistency:")
    cons = decision["consistency"]
    print(f"  Mode          : {cons['consistency_type']}")
    print(f"  Rationale     : {cons['rationale']}")

    print("-" * 60)

    # ✅ Return structured output for experiments / evaluation
    return None
```

```python
sample_text = (
    "Alia Bhatt is a British actress of Indian descent who predominantly works in Hindi films. "
    "Known for her portrayals of women in challenging circumstances, she has received several accolades, "
    "including a National Film Award and seven Filmfare Awards. She is one of India's highest-paid actresses. "
    "Time awarded her with the Time100 Impact Award in 2022 and named her one of the 100 most influential people in the world in 2024."
)

sample_video = "/content/drive/MyDrive/sample_video/test_obama.mp4"

run_c3_mmfd(
    text=sample_text,
    video_path=sample_video
)
```
